```python
import numpy as np
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.show()
```

```python
#Создать объект фигуры с именем fig, используя plt.figure() **
#Используйте add_axes, чтобы добавить ось на холст рисунка в [0,0,1,1]. 
#Назовите эту новую ось топором axis ax.
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
#Создайте объект фигуры и поместите на него две оси, ax1 и ax2. 
#Расположены в [0,0,1,1] и [0,2,0,5,.2,.2] соответственно
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

ax1.plot(x,y, color='red')
ax2.plot(x,y, color='red')
```

```python
#Создайте график добавив две оси к объекту фигуры в [0,0,1,1] и [0,2,0,5,.4,.4]
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.4,.4])

ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')


ax.plot(x,y)
ax.set_title('Zoom')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.4,.4])

ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')


ax2.plot(x,y)
ax2.set_title('Zoom')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
```
